```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
def add_proximity(name, prox_lat, prox_lon, color='#1e90ff'):
    """Add a marker with distance label and a line from the launch site to a proximity point."""
    # distance (KM)
    dist = calculate_distance(launch_site_lat, launch_site_lon, prox_lat, prox_lon)

    # marker with distance text
    folium.Marker(
        [prox_lat, prox_lon],
        icon=DivIcon(
            icon_size=(220, 22),
            icon_anchor=(0, 0),
            html=f'<div style="font-size:12px;color:{color};"><b>{name}: {dist:0.2f} KM</b></div>'
        )
    ).add_to(site_map)

    # line from launch site to the proximity
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [prox_lat, prox_lon]],
        weight=2, color=color
    ).add_to(site_map)

    return dist
```

These patterns line up with operational and safety needs: ocean flight paths, recovery logistics, heavy transport access, and standoff from populated areas.
